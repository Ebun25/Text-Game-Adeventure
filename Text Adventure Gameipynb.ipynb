TEXT ADVENTURE GAME - THE WOMAN KING 
The woman king movie is about an all-female military unit in the army of the Kingdom of Dahomey in southern Benin, on the West coast of Africa, are the subject of The Woman King. The next generation of female recruits are trained by Nanisca, a general famed for her performances in The Help and Fences, to combat enemy tribes that threaten their territory.The Oyo Empire invades the Dahomey Kingdom and kidnaps and sells Dahomey people as slaves. She directs a significant counterattack against the Oyo to defeat their adversaries and free the captured people.
This code creates a text-based adventure game called "Woman King." The game is divided into three stages, and in each stage, the player is presented with a scenario and must make a choice on how to proceed. In the first stage, the player is on a battlefield and must choose to either attack or run. If the player chooses to attack, they are congratulated for saving the kingdom and being a true Woman King.If the player chooses to run, they are scolded for abandoning their duty and not being a true Woman King.In the second stage, the player finds themselves in a forest and must choose to either explore or rest. If the player chooses to explore, they discover the enemy's den and must choose to either hide or fight. If the player chooses to fight, they defeat the enemies and are congratulated for being a real Woman King. In the third stage, the player's warriors have been kidnapped, 
and the king will not authorize a rescue mission. The player must choose whether to defy orders and rescue the captives or not. If the player chooses to rescue the captives, they may be crowned Woman King.

```python
## Text Adventure Game- Woman King

##import random option
import random

##Defining the win function
def win():
    try:
        winner = input(prompt=f"Can you please tell me your name?\n").capitalize()
        print(f"Congrats {winner}, You Won!! You are the True Woman King!!!")
    except:
        print("An error occurred. Please try again.")

# Defining the fail function
def fail():
    try:
        Loosed = input(prompt=f"Can you please tell me your name?\n").capitalize()
        print(f"{Loosed}, You failed!! You cannot be the True Woman King!!!")
    except:
        print("An error occurred. Please try again.")

        
 # Display the introduction of woman king text adventure game
def display_intro():
    try:
        print("Welcome to the game Woman King!\n"
              "You are a member of the all-female military known as Woman King.\n"
              "Your mission is to defeat the enemy and save your kingdom.\n")

# counter for invalid choices
        invalid_choices = 0
        while invalid_choices < 3:
            text = input("Are you ready for the battle? yes/no\n").strip().lower()
            
# Using the if function for the YES/NO response
            if text == 'yes':
                print("Cool, let's begin!\n")
                input("Press Enter to Continue\n")
                print("You are in a battlefield, and you see the enemy approaching.")
                
# player's choice
                choice = input("What do you want to do? (attack/run)\n").strip().lower()
                if choice == "attack":
# player chooses to attack
                    print("You charge towards the enemy and engage in battle.")
                    print("You have defeated the enemy!")
                    rand_num = random.randint(1, 10)
# player fails to loot
                    if rand_num <= 3:
                        fail()
                    else:
# player successfully loots
                        win()
                    break
                elif choice == "run":
# player chooses to run
                    print("You turn and run away from the battlefield.")
                    print("You have abandoned your duty and your people.")
                    print("You are not a true Woman King.")
                    fail()
                    break
                    
# invalid option
                else:
                    print("Invalid choice. Please enter 'attack' or 'run'")
                    invalid_choices += 1
                    if invalid_choices == 3:
                        print("you loose")
                    continue
            elif text == 'no':
                print("Thank you, we love to see you again!")
                break
            else:
                print("Invalid choice. Please enter 'yes' or 'no'")
                invalid_choices += 1
                if invalid_choices == 3:
                    print("you loose")
    except:
        print("An error occurred. Please try again.")
        
# Call the function to display the code
display_intro()


def stage_2():
    try:
##introducing the second stage
# print the introduction message for stage 2
        print("Your kingdom is under threat and you find yourself in a forest.")
        print("However, you are willing to explore and also take a rest")
        print("but you still need to save your people")
    
##counter for invalid choices, while loop and player's choice
# set a counter for invalid choices
        invalid_choices = 0
# start a while loop that runs while the number of invalid choices is less than 3
        while invalid_choices < 3:
# get the player's decision to either explore or rest
            decision = input(prompt =f"What do you want to do? explore/rest\n").strip().lower()
# check if the decision is valid (either explore or rest)
            if decision in ["explore", "rest"]:
# if the decision is valid, print a message and get the player's decision to hide or fight
                print("you start exploring the forest and discovered the enemies den")
                input(prompt=f"<Press Enter to Continue>\n")
                invading = input(prompt=f"Make a decision(hide/fight)\n").strip().lower()
# if the decision to hide, print a message and break out of the while loop
                if invading == "hide":
                    print("You are a coward.")
                    print("Be brave!!")
                    break
# if the decision to fight, print a message and break out of the while loop
                elif invading == "fight":
                    print("You defeated the enemies")
                    print("You've always been a Real Woman king!")
                    print("Keep up the good work!")
                    break
# if the decision is not valid, print an error message and increment the counter for invalid choices
                else:
                    print("Invalid choice. Please enter 'hide' or 'fight'.")
                    invalid_choices += 1
# if the decision is not valid, print an error message and increment the counter for invalid choices
            else:
                print("Invalid choice. Please enter 'explore' or 'rest'")
                invalid_choices += 1
# if the player has entered too many invalid choices, print an error message and break out of the while loop
                if invalid_choices == 3:
                    print("You have entered too many invalid choices.")
# if an error occurs, print an error message
    except:
        print("An error occurred. Please try again.")
        
# Call the function to test it
stage_2()


import random

def stage_3():
# Display introduction
    print("""You realized your warriors have been kidnapped but the king will 
          not authorize a rescue mission for the captives. 
          You decided to defile orders and sets out with a 
          group of like-minded warriors to rescue the captives.\n""" )

# Set up prompts
    prompts = [
        "Time to rescue!!!(Are you in/out?)\n",
        "Are you in for this challenge? (in/out)\n"
    ]

# Select a random prompt
    prompt = random.choice(prompts)

# Create the while loop and player's choice
    trials = 0
    while trials < 2:  # loop while the number of trials is less than 2
        try:
            rescue_mission = input(prompt=prompt).strip().lower()  # get user input, strip any leading/trailing spaces, and convert to lowercase
            if rescue_mission == 'in':  # if the player chooses 'in'
                print("You are always dedicated to the kingdom")
                print("You might be crowned the Woman King")
                break  # break out of the loop
            elif rescue_mission == 'out':  # if the player chooses 'out'
                print("You keep acting like a spy")
                print("Prove that you are not a spy")
                trials += 1  # increase the number of trials
                if trials == 2:
                    print("Bye")
            else:  # if the player enters any other input
                print("Invalid choice. Please enter 'in' or 'out'.")
                trials += 1  # increase the number of trials
                if trials == 2:
                    print("Bye")
        except:  # if an error occurs during input
            print("An error occurred. Please try again.")

# Call the function to test it
stage_3()


def stage_4():
    ## Display the introduction
    print("""You are to be crowned the woman king but the crown is missing. 
         You and your all-female military started a search for it . 
         There are three places to be checked "underground", "forest" and an "abandoned building".
         I hope you find it because you truly deserve to be called the woman king""")

# counter for invalid choices
    trials = 0
    
# loop until player wins or exhausts all attempts
    while trials < 3:
        try:
# prompt player for answer
            answer_attempt = input(prompt=f"Where could the crown be?\n. underground, forest, or abandoned building: ")
            
# check if player's answer is correct
            if answer_attempt == 'underground':
                print("Congratulations")
                print("You found the crown")
                print("You are now the Woman King!")
                win()
                break
            
# check if player's answer is invalid
            elif answer_attempt not in ['underground', 'forest', 'abandoned building']:
                print("Invalid choice. Please enter 'underground' or 'forest' or 'abandoned building'.")
                trials += 1
            
# check if player's answer is incorrect
            elif answer_attempt == 'forest':
                print("No one kept the crown here")
                print("Your search wasn't right")
                print("You cannot be crowned the woman king!")
                trials += 1
                
            elif answer_attempt == "abandoned building":
                print("I am tired")
                print("The crown isn't here")
                print("I do not deserve to be called the woman king")
                trials += 1
            else:
                print('\nIncorrect answer.\n')
                trials += 1
                
# check if the player has exhausted all attempts
                if trials == 3:
                    print("you lose")
                    fail()
        except:
# handle any exceptions that may occur during the input process
            print("An error occurred. Please try again.")

stage_4()
```
